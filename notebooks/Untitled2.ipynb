```python
def psi(x, x0=0.4, delta=0.33):
    """
    Gaussian function
    Input:
    ======
    x: float or numpy array
    
    x0: optional, float, the centre of the distribution
    
    delta: optional, float, the with of the distribution
    
    Output:
    =======
    float or numpy array:, The value of the function at x
    
    Example:
    ========
    >>> psi(0.4)
    0.57990277133688173
    
    >>> psi(0.4, delta=0.4)
    0.60492681129785841
    
    """
    from numpy import exp,sqrt, pi
    return 1/sqrt(2*pi)/delta*exp(-1/2*((x-x0)/delta)**2)
```

```python
print("The value of psi({:.1f}) is {:.4f}".format(0.8,psi(0.8)))
print("The value of psi({:.1f}) is {:.4f}".format(0.8,psi(0.8, 0.4, 0.33)))
print("The value of psi({:.1f}) is {:.4f}".format(0.8,psi(0.8, 0.5, 0.2)))
```
